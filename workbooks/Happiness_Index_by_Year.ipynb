```python
import pandas as pd 
import numpy as np
```

```python
#bring in happiness files 
h2015= pd.read_csv("../Data/happiness2015.csv")
h2016= pd.read_csv("../Data/happiness2016.csv")
h2017= pd.read_csv("../Data/happiness2017.csv")
h2018= pd.read_csv("../Data/happiness2018.csv")
h2019= pd.read_csv("../Data/happiness2019.csv")
```

```python
happiness1 = pd.merge(h2015, h2016, on='Country')
happiness1
```

```python
happiness2 = pd.merge(happiness1, h2017, on='Country')
happiness2
```

```python
happiness3 = pd.merge(happiness2, h2018, on='Country')
happiness3
```

```python
happiness4 = pd.merge(happiness3, h2019, on='Country')
happiness4
```

```python
gdp_pop= pd.read_csv("../Data/Cleaned_data/Suicides_country_GDP_Population.csv")
gdp_pop
```

```python
happiness = pd.merge(happiness4, gdp_pop, on='Country')
happiness
```

```python
#convert to CSV
happiness.to_csv('Happiness_index.csv')
```
